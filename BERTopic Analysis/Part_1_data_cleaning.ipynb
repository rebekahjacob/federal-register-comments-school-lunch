```python
import pandas as pd
import numpy as np

df_2017=pd.read_csv("data/raw_data/first-round-comments.csv" , index_col=False)
df_2020=pd.read_csv("data/raw_data/second-round-comments.csv", index_col=False)
```

```python
df_2017=df_2017.drop(columns=["Unnamed: 0","stance_1","stance_2"])
df_2020=df_2020.drop(columns=["Unnamed: 0","stance_1","stance_2"])
df_2017["form_letter"]=df_2017["form_letter"].astype(str)
```

```python
set(list(df_2020["stance_1_cat3"]))
```

```python
# List of substrings to check
patterns = ['AHA Form', 'Salud', 'SNA Form',"Latino Form"]

# Filter out rows where 'form_letter' starts with any of the patterns
df_2017 = df_2017[~df_2017['form_letter'].str.startswith(tuple(patterns), na=False)]
df_2017
```

```python
set(list(df_2017["form_letter"]))
```

```python
# List of substrings to check
patterns = [ "Form 1a", "Form 2- Salud America", "Form 3"]

# Filter out rows where 'form_letter' starts with any of the patterns
df_2020 = df_2020[~df_2020['form_letter'].str.startswith(tuple(patterns), na=False)]
df_2020
```

```python
set(list(df_2020["form_letter"]))
```

```python
df=pd.concat([df_2017,df_2020]).reset_index(drop=True)
df
```

```python
set(list(df["stance_1_cat3"]))
```

```python
df_flexibilities=df[df["stance_1_cat3"]=='In favor of flexibilies'].reset_index(drop=True)
df_flexibilities.to_csv("data/interim_data/df_flexibilities.csv", index=False)
```

```python
df_HHFKA=df[df["stance_1_cat3"]=='In favor of original HHFKA'].reset_index(drop=True)
df_HHFKA.to_csv("data/interim_data/df_HHFKA.csv", index=False)
```
